```python
import h3
import pandas as pd
import geopandas as gpd
import geodatasets
```

```python
path = geodatasets.get_path("naturalearth.land")

world = gpd.read_file(path)
world = gpd.read_file(
    "https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip"
)

brazil = world[world["ADMIN"] == "Brazil"].geometry.iloc[0]

coords = list(brazil.exterior.coords)
polygon = h3.LatLngPoly(
    [(lat, lng) for lng, lat in coords]
)

res5_cells = list(
    h3.h3shape_to_cells(
        polygon,
        res=5
    )
)
```

```python
data = {
    "res5": res5_cells
}

for res in range(4, -1, -1):
    data[f"res{res}"] = [
        h3.cell_to_parent(h, res)
        for h in res5_cells
    ]

df = pd.DataFrame(data)
df
```

```python
coords = df["res5"].apply(h3.cell_to_latlng)

df["lat"] = coords.str[0]
df["lng"] = coords.str[1]

df
```

```python
for res in range(5, -1, -1):
    df[f'res{res}_idx'] = pd.factorize(df[f'res{res}'])[0]

df
```

```python
for res in range(5, -1, -1):
    df[f'res{res}_cor'] = (df[f'res{res}_idx'] % 15) + 1

df
```

```python
cont = 0

for res in range(5, -1, -1):
    df_aux = df[['lat','lng']]
    df_aux['resolucao'] = res
    df_aux['cor'] = df[f"res{res}_cor"]

    if cont == 0:
        df2 = df_aux.copy()
    else:
        df2 = pd.concat([df2, df_aux])

    cont = cont + 1

df2
```

```python
df2.to_csv('carga.csv',index=False)
```
