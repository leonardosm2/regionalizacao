```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from sklearn.cluster import KMeans
import numpy as np
```

```python
gdf = gpd.read_file("BR_setores_CD2022.shp")
```

```python
ponto = Point(-46.6333, -23.5505)
resultado = gdf[gdf.contains(ponto)]
resultado['SITUACAO']
```

```python
gdf['lon'] = gdf['geometry'].apply(lambda g: g.centroid.x)
gdf['lat'] = gdf['geometry'].apply(lambda g: g.centroid.y)
gdf
```

```python
gdf.drop('geometry', axis=1, inplace=True)
```

```python
df = pd.DataFrame(gdf)
df
```

```python
df['score'] = 1
df.loc[df['SITUACAO']=='Urbana', 'score'] = 2
df['city'] = df['NM_MUN']
df['uf'] = df['NM_UF']
df['ramo'] = 'ALL'

export = df[['city','lat','lon','score','ramo','uf']]
export
```

```python
coords = np.radians(export[['lat', 'lon']])

kmeans = KMeans(n_clusters=7000, random_state=42)
export['cluster'] = kmeans.fit_predict(coords)

export
```

```python
score = export[['cluster','score']].groupby('cluster').mean().reset_index().rename(columns={'score':'score_medio'})
score['score'] = pd.qcut(
    score['score_medio'],
    q=4,
    labels=[1, 2, 3, 4]
)
score.drop(columns='score_medio', inplace=True)
score
```

```python
export = pd.merge(export, score, on='cluster', how='inner').drop_duplicates(subset='cluster', keep='first')
export['score_x'] = export['score_y']
export.drop(columns=['cluster', 'score_y'], inplace=True)
export.rename(columns={'score_x': 'score'}, inplace=True)
export
```

```python
export.to_csv('mapinha.csv',index=False)
```
